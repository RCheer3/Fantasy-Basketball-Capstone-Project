```python
parameters = {'learning_rate':[0.01,0.05,0.1], 'n_estimators':[500,1000,2000],'max_depth':[3,5,7],'max_features':[0.3,0.5]}

clf_assists = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_assists.fit(cat_data.X_train_assists, cat_data.y_train_assists)

print('GB Assists r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_assists.best_params_['learning_rate'],max_depth = clf_assists.best_params_['max_depth'],
                         max_features=clf_assists.best_params_['max_features'],n_estimators=clf_assists.best_params_['n_estimators'])
                ,cat_data.X_train_assists,cat_data.y_train_assists,cv=5))))
print('LR Assists r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_assists,cat_data.y_train_assists,cv=5))))



clf_rebounds = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_rebounds.fit(cat_data.X_train_rebounds, cat_data.y_train_rebounds)

print('GB Rebounds r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_rebounds.best_params_['learning_rate'],max_depth = clf_rebounds.best_params_['max_depth'],
                         max_features=clf_rebounds.best_params_['max_features'],n_estimators=clf_rebounds.best_params_['n_estimators'])
                ,cat_data.X_train_rebounds,cat_data.y_train_rebounds,cv=5))))
print('LR Rebounds r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_rebounds,cat_data.y_train_rebounds,cv=5))))



clf_steals = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_steals.fit(cat_data.X_train_steals, cat_data.y_train_steals)

print('GB Steals r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_steals.best_params_['learning_rate'],max_depth = clf_steals.best_params_['max_depth'],
                         max_features=clf_steals.best_params_['max_features'],n_estimators=clf_steals.best_params_['n_estimators'])
                ,cat_data.X_train_steals,cat_data.y_train_steals,cv=5))))
print('LR Steals r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_steals,cat_data.y_train_steals,cv=5))))



clf_blocks = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_blocks.fit(cat_data.X_train_blocks, cat_data.y_train_blocks)

print('GB Blocks r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_blocks.best_params_['learning_rate'],max_depth = clf_blocks.best_params_['max_depth'],
                         max_features=clf_blocks.best_params_['max_features'],n_estimators=clf_blocks.best_params_['n_estimators'])
                ,cat_data.X_train_blocks,cat_data.y_train_blocks,cv=5))))
print('LR Blocks r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_blocks,cat_data.y_train_blocks,cv=5))))


clf_threes = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_threes.fit(cat_data.X_train_threes, cat_data.y_train_threes)

print('GB Threes r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_threes.best_params_['learning_rate'],max_depth = clf_threes.best_params_['max_depth'],
                         max_features=clf_threes.best_params_['max_features'],n_estimators=clf_threes.best_params_['n_estimators'])
                ,cat_data.X_train_threes,cat_data.y_train_threes,cv=5))))
print('LR Threes r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_threes,cat_data.y_train_threes,cv=5))))


clf_tos = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_tos.fit(cat_data.X_train_tos, cat_data.y_train_tos)

print('GB Tos r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_tos.best_params_['learning_rate'],max_depth = clf_tos.best_params_['max_depth'],
                         max_features=clf_tos.best_params_['max_features'],n_estimators=clf_tos.best_params_['n_estimators'])
                ,cat_data.X_train_tos,cat_data.y_train_tos,cv=5))))
print('LR Tos r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_tos,cat_data.y_train_tos,cv=5))))

clf_FGM = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_FGM.fit(cat_data.X_train_FGM, cat_data.y_train_FGM)

print('GB FGM r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_FGM.best_params_['learning_rate'],max_depth = clf_FGM.best_params_['max_depth'],
                         max_features=clf_FGM.best_params_['max_features'],n_estimators=clf_FGM.best_params_['n_estimators'])
                ,cat_data.X_train_FGM,cat_data.y_train_FGM,cv=5))))
print('LR FGM r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_FGM,cat_data.y_train_FGM,cv=5))))

parameters = {'learning_rate':[0.01,0.05,0.1], 'n_estimators':[500,1000,2000],'max_depth':[3,5,7],'max_features':[0.3,0.5]}

clf_FGP = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_FGP.fit(cat_data.X_train_FGP, cat_data.y_train_FGP)

print('GB FGP r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_FGP.best_params_['learning_rate'],max_depth = clf_FGP.best_params_['max_depth'],
                         max_features=clf_FGP.best_params_['max_features'],n_estimators=clf_FGP.best_params_['n_estimators'])
                ,cat_data.X_train_FGP,cat_data.y_train_FGP,cv=5))))
print('LR FGP r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_FGP,cat_data.y_train_FGP,cv=5))))

clf_FTM = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_FTM.fit(cat_data.X_train_FTM, cat_data.y_train_FTM)

print('GB FTM r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_FTM.best_params_['learning_rate'],max_depth = clf_FTM.best_params_['max_depth'],
                         max_features=clf_FTM.best_params_['max_features'],n_estimators=clf_FTM.best_params_['n_estimators'])
                ,cat_data.X_train_FTM,cat_data.y_train_FTM,cv=5))))
print('LR FTM r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_FTM,cat_data.y_train_FTM,cv=5))))


clf_FTP = GridSearchCV(GradientBoostingRegressor(), parameters, cv=5,n_jobs=-1,verbose=3)
clf_FTP.fit(cat_data.X_train_FTP, cat_data.y_train_FTP)

print('GB FTP r2: {}'.format(np.mean(cross_val_score(GradientBoostingRegressor(learning_rate=clf_FTP.best_params_['learning_rate'],max_depth = clf_FTP.best_params_['max_depth'],
                         max_features=clf_FTP.best_params_['max_features'],n_estimators=clf_FTP.best_params_['n_estimators'])
                ,cat_data.X_train_FTP,cat_data.y_train_FTP,cv=5))))
print('LR FTP r2: {}'.format(np.mean(cross_val_score(LinearRegression(fit_intercept=True)
                ,cat_data.X_train_FTP,cat_data.y_train_FTP,cv=5))))

```
